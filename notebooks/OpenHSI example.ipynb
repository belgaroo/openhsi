```python
import matplotlib.pyplot as plt
import numpy as np
import pyfits
%matplotlib inline

from openhsi import openhsi
```

```python
ohsi=openhsi()
ohsi.start()
```

```python
ohsi.exposure=1000
ohsi.exposure
```

```python
img=np.rot90(ohsi.get_single_image(),-1)
plt.figure(figsize=(15,15))
imgplot = plt.imshow(img)
```

```python
cube=ohsi.get_image_series(numimages=100)
```

```python
pyfits.writeto("cube.fits", np.mean(cube, axis=2), clobber=True)
```
